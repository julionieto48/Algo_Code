Two sums

```python
nums = [4,5,6,3,5,8]
target = 10

for i in range(len(nums)):

  for j in range(i+1, len(nums)):
    if nums[i] + nums[j] == target:
      print(i, j)

```

```python
nums = [4,5,6,3,5,8]
target = 10
visitado = []

for i in range(len(nums)):
  visitado.append(nums[i])
  faltante = target - nums[i]
  if (faltante in visitado) and (faltante != nums[i] ) :
    print(nums[i], faltante)
```

```python
nums = [4,5,6,3,5,8]
target = 10

vistos = {}

for i in range(len(nums)):
  actual = nums[i]
  faltante = target - actual
  if faltante in vistos:
    print("Índices:", [vistos[faltante], i])

  # Si no, guardo el número con su posición
  vistos[actual] = i
```

```python
nums = [2, 7, 4, 0, 9, 5, 1, 3]
target = 6

for i in range(len(nums)):
    for j in range(i+1, len(nums)):
        for k in range(j+1, len(nums)):
            if nums[i] + nums[j] + nums[k] == target:
                print(nums[i], nums[j], nums[k])
```
